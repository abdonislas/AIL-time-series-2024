```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import datetime as dt
import seaborn as sns
from statsmodels.graphics.tsaplots import month_plot
```

```python
stock = "NVDA"
ticker = yf.Ticker(stock)
data = ticker.history(start = "2013-01-01", end = "2023-12-31", interval = "1mo")["Close"]
data.head()
```

```python
sns.lineplot(data = data)
plt.ylabel("Precio")
plt.xlabel("Fecha")
plt.title("Nvidia")
```

```python
# Creamos una nueva columna 'year' y 'month' en el Ã­ndice
data = data.reset_index()
data['year'] = data['Date'].dt.year
data['month'] = data['Date'].dt.month

# Creamos la tabla pivote
piv = pd.pivot_table(data, values='Close', index='year', columns='month')
piv.plot(figsize=(12,8))
plt.legend(pd.unique(data['month']))
plt.show()
```
