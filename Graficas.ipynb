```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


nvda = yf.Ticker("NVDA")
data_nvda = nvda.history(period="max")
serie_nvda = pd.Series(data_nvda['Close'], index=data_nvda.index)


gold = yf.Ticker("GLD")
data_gold = gold.history(period="max")
serie_gold = pd.Series(data_gold['Close'], index=data_gold.index)


fig, ax1 = plt.subplots(figsize=(12, 6))


ax1.plot(serie_nvda.index, serie_nvda.values, label='NVIDIA', color='blue')
ax1.set_xlabel('Fecha')
ax1.set_ylabel('Precio NVIDIA ($)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')


ax2 = ax1.twinx()
ax2.plot(serie_gold.index, serie_gold.values, label='Oro (GLD)', color='gold')
ax2.set_ylabel('Precio Oro ($)', color='gold')
ax2.tick_params(axis='y', labelcolor='gold')

plt.title('Comparacin de precios histricos: NVIDIA vs Oro')
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))


plt.tight_layout()
plt.show()
```
